## Instalación de librerías de R

Si necesitamos una librería adicional podemos añadirla con `install.packages`. Por ejemplo, descomentando la siguiente línea podemos asegurarnos de que se instala el `tidyverse`:

```r
# install.packages("tidyverse")
install.packages("stringr")
```

Y ahora podemos utilizarlas en la forma habitual:

```r
library(tidyverse)

cortes = seq(min(mpg$cty), max(mpg$cty), length.out = 11)

ggplot(data = mpg) + 
    geom_histogram(mapping = aes(cty), breaks = cortes, 
               fill = "orange", color="black")
```

## Ejemplo de separate

```r
set.seed(2019)
codigo = paste0(sample(1:5, 6, replace = TRUE), "/", sample(LETTERS[1:2], 6, replace = TRUE))
(datos = data.frame(x = sample(1:10, 6, replace = TRUE), codigo))

datos %>% 
    separate("codigo", into = c("Numero", "Letra"))

```
